```python
import matplotlib.pyplot as plt
```

```python
t = [0,1]
x = 234
y = 213

s = (y/x - 1) * 100

print(f'The percent change of the stock price is {s:.2f}%.')
```

```python
#plt.scatter(t, [x,y])
plt.plot(t, [x, y], linestyle="--", color = 'purple', linewidth = 2, marker="o", label = f'Percent change: {s:.2f}%')

plt.xlabel('Time')
plt.ylabel('Value')

plt.legend()
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
import yfinance as yf

aapl = yf.Ticker('AAPL')
data = aapl.history(period = '1y')
print(data.head())
```

```python
plt.plot(data.index, data['Close'])
plt.tight_layout()
plt.show()
```

```python
data['Percent_change_1'] = data['Close'].pct_change() * 100
data['Percent_change_2'] = (data['Close'] / data['Close'].shift(1) - 1) * 100
data.dropna(inplace = True)
data.head()
```

```python
plt.plot(data.index, data['Percent_change_1'])
plt.tight_layout()
plt.show()
```
