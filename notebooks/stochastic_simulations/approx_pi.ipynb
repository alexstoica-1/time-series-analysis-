# Example of basic MC Simulation

```python
import pandas as pd
import numpy as np
import scipy
import matplotlib.pyplot as plt
np.random.seed(42)
```

### Approximating Pi

```python
n_simulations = 20000

xs = np.random.uniform(-1, 1, size=n_simulations)
ys = np.random.uniform(-1, 1, size=n_simulations)

# boolean mask
inside = (xs**2 + ys**2) <= 1
n_points_circle = int(inside.sum())
n_points_square = n_simulations
pi = 4 * n_points_circle / n_points_square

plt.figure(figsize=(6,6))

# with ~inside you just invert the boolean mask array
plt.scatter(xs[~inside], ys[~inside], s=1, color='red', alpha=0.5, label='outside')
plt.scatter(xs[inside], ys[inside], s=1, color='blue', alpha=0.5, label='inside')

theta = np.linspace(0, 2*np.pi, 400)
plt.plot(np.cos(theta), np.sin(theta), color='black', linewidth=1)
ax = plt.gca()
ax.set_aspect('equal', 'box')
plt.title(f'Monte Carlo Pi approximation ({n_simulations} samples)')
plt.grid(False)
plt.text(0, -1.06, f"Approx value of  is {pi:.5f}.", 
         fontsize=10, ha='center', va='center', fontweight='bold')
plt.show()
```

```python
print(f"The approximated value of pi from the experiment is: {pi:.5f}.")
```
