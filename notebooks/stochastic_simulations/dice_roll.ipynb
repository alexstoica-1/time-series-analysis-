# Dice Roll

```python
import numpy as np
import random
import matplotlib.pyplot as plt
random.seed(42)
```

```python
def roll_dice(n, seed):
    np.random.seed(seed)
    total_dice = 0
    point_dice = []
    for i in range(n):
        total_dice += 1
        point_dice.append(np.random.randint(1,7))

    mean_point = np.mean(point_dice)

    return([total_dice, float(mean_point)])
```

```python
seed = 42
for i in [10,100,1000,10000]:
    x = roll_dice(i,seed)
    print(f"The expected value that we get for {x[0]} rolls is: {x[1]}.")
```

```python
# two bags of biased dice
bag1 = [[1,2,3,6,6,6],[1,2,3,4,4,6],[1,2,3,3,3,5]]
bag2 = [[2,2,3,4,5,6],[3,3,3,4,4,5],[1,1,2,3,5,5]]
```

```python
def roll_biased_dice(n):
    results = {}
    for i in range(n):
        bag_index1 = random.randint(0,2)
        die_index1 = random.randint(0,5)

        bag_index2 = random.randint(0,2)
        die_index2 = random.randint(0,5)

        point1 = bag1[bag_index1][die_index1]
        point2 = bag2[bag_index2][die_index2]

        key = "%s_%s" % (point1, point2)

        if point1 + point2 == 8:
    
            if key not in results:
                results[key] = 1
            else: 
                results[key] += 1
                
    return results

x = roll_biased_dice(1000)
print(x)
```

```python
colors = [plt.cm.tab10(i) for i in random.sample(range(10), len(x))]

plt.bar(x.keys(), x.values(), color = colors)
plt.xlabel('Two dice')
plt.ylabel('Frequency')
plt.title('Visualization of random draws')
plt.show()
```
