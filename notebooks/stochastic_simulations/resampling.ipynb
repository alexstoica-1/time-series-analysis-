```python
import numpy as np
import random
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
def two_random_states():
    ne_states = ['Maine',
                'Vermont',
                'New Hampshire',
                'Massachusestts']

    return(random.sample(ne_states, 2))

two_random_states()
```

```python
# Bootstrapping 
# You randomly sample n observations from your sample, compute a statistic with the smaller sample, then repeat the process many times

nba_heights = [196, 191, 216, 188, 185, 211, 201, 208, 191, 196, 165, 199, 189, 192,193]
simu_heights = []

for i in range(1000):
    bootstrap_sample = random.choices(nba_heights, k = 10)
    simu_heights.append(np.mean(bootstrap_sample))

upper = np.quantile(simu_heights, 0.975)
lower = np.quantile(simu_heights, 0.025) # alpha = 0.05, quantile is alpha/2
print([f"{np.mean(simu_heights):.2f}", float(lower), float(upper)])
```

```python
sns.displot(simu_heights, stat="density", kde = True)
plt.axvline(float(lower), color = 'red')
plt.axvline(float(upper), color = 'red')
plt.axvline(float(np.mean(simu_heights)), color = 'green')

plt.grid(True)
plt.show()
```
