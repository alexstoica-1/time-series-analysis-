# **Generator** from Numpy

```python
import numpy as np
```

```python
# relies on BitGenerator to manage state and generate random bits
randomizer = np.random.default_rng(seed = 42)

rand_nb = randomizer.random()
print(rand_nb, type(rand_nb))
```

```python
rand_ints = randomizer.integers(low = 0, high = 10,size = 11)
print(rand_ints, type(rand_ints[0]), type(rand_ints)) 
```

```python
from numpy.random import Generator, PCG64
import matplotlib.pyplot as plt

rng = Generator(PCG64())
normal = rng.standard_exponential(size = 20)
#print(normal)

plt.plot(normal, marker = '*')
plt.title("Random Normal Samples")
plt.ylabel('Index')
plt.xlabel('Value')
plt.grid(True)
plt.show()
```

```python
big_normal = rng.standard_normal(size = 10000)

plt.hist(big_normal, bins=20, density=True, alpha=0.7, color='green', edgecolor='black')
x = np.linspace(-4, 4, 200) # evenly spaced numbers in the interval
pdf = (1 / np.sqrt(2 * np.pi)) * np.exp(-0.5 * x**2) # plot them over the normal curve 
plt.plot(x, pdf, 'r-', linewidth=2, label='Normal PDF')
plt.title('Histogram of Standard Normal Samples with Normal Curve')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(False)
plt.show()
```
