# Generating Discrete Random Variables

```python
import numpy as np
import scipy.stats as st
import seaborn as sns 
import pandas as pd 
import matplotlib.pyplot as plt 
import random 

random.seed(42)
np.random.seed(42)
```

```python
"""
One can also create a reproducible random generator manually and the use it in Scipy generation.

rng = np.random.default_rng(42)
sample = st.norm.rvs(size = ..., random_state=rng)


"""
```

```python
low = 2
high = 20
samples = st.randint.rvs(low, high, size = 1000) # random variates sampling
samples_dict = {'nums': samples}
ax = sns.histplot(x = "nums", data = samples_dict, color = 'green', bins = 6, binwidth = 0.5)
ax.set_title(f"Uniform distribution [{low},{high}]")
ax.set_ylabel("Number of samples")
ax.set_xlabel("Number from distribution")
plt.show()
```

```python
proba = 0.6
samples = st.geom.rvs(proba, size = 10000)
samples_dict = {'nums': samples}
ax = sns.histplot(x = 'nums', data = samples_dict, color = 'green', binwidth=1, edgecolor='black')
ax.set_xticks(range(1,16))
ax.set_ylabel("Number of samples")
ax.set_xlabel("Number from distribution")
ax.set_title(f"Geometric Distribution, p = {proba}")
plt.show()
```

# Generating Continuous Random Variables 

```python
heights = st.norm.rvs(loc = 177, scale = 8, size = 10000) # loc is mean and scale is std

qualified = (heights < 165) | (heights > 190)

print(np.sum(qualified)*  100 / 10000)

heights_dict = {'heights' : heights}
mean_value = np.mean(heights)

ax = sns.histplot(x = 'heights', data = heights_dict, color = 'green', binwidth = 0.6)
plt.axvline(x = mean_value, color = 'blue')
plt.axvline(x = 165, color = 'red')
plt.axvline(x = 190, color = 'red')
#plt.text(mean_value, 0, 'mean', color='blue', ha='center', va='bottom', fontsize=9)
ax.set_xlabel("Heights")
plt.show()
```

```python
n = 50 # number of flips of the unfair coin 
p = [0.2, 0.8]
size = 500 # number of trials 

results = st.multinomial.rvs(n,p, size = size)
print(type(results), results.shape)

df = pd.DataFrame(
    {
        "Head":results[:,0],
        "Tail":results[:,1]
    }
)
sns.pairplot(df)
plt.show()
```

```python
mean = [2, 6]
size = 500 
cov_mat = np.array([[1,-0.9], [-0.9,1]])

results = st.multivariate_normal.rvs(mean = mean, size = size, cov = cov_mat)
print(type(results), results.shape)

df2 = pd.DataFrame(
    {
        "Price":results[:,0],
        "Demand":results[:,1]
    }
)
sns.pairplot(df2)
plt.show()
```
